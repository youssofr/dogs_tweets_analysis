# Puppos Or Floofers? Which Is Cutest? 

Ever wondered what dog breed attracts the most attention and is popular with most people? Or in what stage of development are dogs the most adorable and cutest? 

To answer these questions we have gathered ifnormation from tweets from [WeRateDogs](https://twitter.com/dog_rates) between 2017 and 2015, and extacted some of the information these tweets could offer from about different stages in dogs' lives and how they appeal to people in each. We have also fed the pictures from these tweets to a machine learning algorithm which classified these images, detected which of them were dog pictures and which were only screnshots of other tweets or anything else, and then predicted dog breeds in each picture. After gathering this data, performing some necessary assessment in order to learn what issues might hinder the analysis or affect prespetice results, and cleaning it from all these issues, our data was ready to tell its answers (and we were so eager to listen, hope you too are : ) !).

We have set the number of favorits and retweets recieved by a certain breed or stage as a measure of its popularity. We have also put in consideration the rating they gave to each dog they tweeted about (otherwise it would be unfair to bypass all their efforts, don't you think! After All "TheyRateDogs"!). So when we talk about most popular and most appealing here you should understand that we mean is what these metrics reflect.
